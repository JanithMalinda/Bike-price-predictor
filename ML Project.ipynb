# Model Selection

```python
np.random.seed(42)
```

```python
x = data.drop('price', axis=1)
x.head(2)
```

```python
x.shape
```

```python
y = data['price']
y.head()
```

```python
#Split The Dataset

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

#checking the dimentions of the train and test dataset
print(f"x_train:{x_train.shape} , x_test:{x_test.shape} y_train:{y_train.shape} , y_test:{y_test.shape}")
```

```python
#Train the model using Linear Regression

from sklearn.linear_model import LinearRegression

model = LinearRegression()

#fit the model
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
#EVALUATING THE MODEL

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

mse = mean_squared_error(y_test, y_pred)
r2Score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("Mean Squared Error: ", mse)
print("R Squared Error: ", r2Score)
print("Mean Absolute Error: ", mae)
print("R2 score for training: ", model.score(x_train,y_train))
print(model.score(x_test, y_test))
```

```python
#visualizing the actual vs predicted values

plt.scatter(y_pred, y_test, color='blue')
plt.xlabel("Predicted values")
plt.ylabel("Actual values")
plt.title("Actual Vs Predicted Values")
plt.show()
```

```python
#Create model Using Random Forest Regressor
```

```python
from sklearn.ensemble import RandomForestRegressor #applying the random forest regressor

clf = RandomForestRegressor()

clf.fit(x_train, y_train)
```

```python
y2_pred = clf.predict(x_test)
```

```python
#Evaluating RandomForestRegressor
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

mse = mean_squared_error(y_test, y2_pred)
r2Score = r2_score(y_test, y2_pred)
mae = mean_absolute_error(y_test, y2_pred)

print("Mean Squared Error: ", mse)
print("R Squared Error: ", r2Score)
print("Mean Absolute Error: ", mae)
print("R2 score for training: ", clf.score(x_train,y_train))
print("R2 score for testing: ", clf.score(x_test,y_test))

clf.score(x_train,y_train)
```

```python
plt.scatter(y_test, y2_pred, color = 'red')

plt.xlabel("Predicted values")
plt.ylabel("Actual values")
plt.title("Actual Vs Predicted Values")
plt.show()
```
